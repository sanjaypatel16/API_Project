```python
# Import Libraries
import requests
import pandas as pd
import matplotlib.pyplot as plt

# Optional: Configure matplotlib for inline display in Jupyter
%matplotlib inline
```

## Understand the Open-Meteo API

- **API Endpoint**: `https://archive-api.open-meteo.com/v1/archive`
- **Parameters**:
  - `latitude`: Latitude of the location.
  - `longitude`: Longitude of the location.
  - `start_date`: Start date of the data (YYYY-MM-DD).
  - `end_date`: End date of the data (YYYY-MM-DD).
  - `hourly`: Comma-separated list of variables (e.g., temperature, precipitation).
  - `timezone`: Timezone of the data.

**Documentation**: [Open-Meteo API Docs](https://open-meteo.com/en/docs)

```python
# Step 1: Set Up Parameters

# Location coordinates for New York City
latitude = 40.7128
longitude = -74.0060

# Date range
start_date = '2023-01-01'
end_date = '2023-01-07'  # One week of data

# Variables to fetch
hourly_variables = ['temperature_2m', 'relativehumidity_2m', 'precipitation']
```

```python
# Step 2: Make the API Request

base_url = 'https://archive-api.open-meteo.com/v1/archive'

params = {
    'latitude': latitude,
    'longitude': longitude,
    'start_date': start_date,
    'end_date': end_date,
    'hourly': ','.join(hourly_variables),
    'timezone': 'America/New_York'
}

response = requests.get(base_url, params=params)
```

```python
# Check if the request was successful

if response.status_code == 200:
    print('Data fetched successfully!')
else:
    print(f'Failed to fetch data. Status code: {response.status_code}')
```

```python
# Step 3: Load Data into pandas DataFrame

data = response.json()
hourly_data = data['hourly']

df = pd.DataFrame(hourly_data)
df.head()
```

```python
# Step 4: Data Cleaning

# Convert 'time' column to datetime
df['time'] = pd.to_datetime(df['time'])

# Set 'time' as the index
df.set_index('time', inplace=True)

# Check for missing values
df.isnull().sum()
```

```python
# Handle missing values (if any)
df.fillna(method='ffill', inplace=True)  # Forward fill
```

```python
# Step 5: Exploratory Data Analysis

# Summary Statistics
df.describe()
```

```python
# Plot Temperature Over Time

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['temperature_2m'], label='Temperature (°C)')
plt.title('Temperature Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.show()
```

```python
# Plot Relative Humidity Over Time

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['relativehumidity_2m'], color='orange', label='Relative Humidity (%)')
plt.title('Relative Humidity Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Relative Humidity (%)')
plt.legend()
plt.show()
```

```python
# Plot Precipitation Over Time

plt.figure(figsize=(14, 6))
plt.bar(df.index, df['precipitation'], label='Precipitation (mm)')
plt.title('Precipitation Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Precipitation (mm)')
plt.legend()
plt.show()
```

```python
# Step 6: Correlation Analysis

# Calculate correlation between temperature and humidity
correlation = df['temperature_2m'].corr(df['relativehumidity_2m'])
print(f'Correlation between Temperature and Relative Humidity: {correlation:.2f}')
```

```python
# Scatter Plot of Temperature vs. Relative Humidity

plt.figure(figsize=(8, 6))
plt.scatter(df['temperature_2m'], df['relativehumidity_2m'], alpha=0.5)
plt.title('Temperature vs. Relative Humidity')
plt.xlabel('Temperature (°C)')
plt.ylabel('Relative Humidity (%)')
plt.show()
```

```python
# Step 7: Resample Data to Daily Averages

daily_avg = df.resample('D').mean()
daily_avg.head()
```

```python
# Plot Daily Average Temperature

plt.figure(figsize=(10, 5))
plt.plot(daily_avg.index, daily_avg['temperature_2m'], marker='o')
plt.title('Daily Average Temperature in New York City')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()
```

```python
# Step 8: Compare with Another Location (Los Angeles)

# Coordinates for Los Angeles
latitude_la = 34.0522
longitude_la = -118.2437

params_la = {
    'latitude': latitude_la,
    'longitude': longitude_la,
    'start_date': start_date,
    'end_date': end_date,
    'hourly': ','.join(hourly_variables),
    'timezone': 'America/Los_Angeles'
}

response_la = requests.get(base_url, params=params_la)

# Load and clean Los Angeles data
data_la = response_la.json()
hourly_data_la = data_la['hourly']
df_la = pd.DataFrame(hourly_data_la)

df_la['time'] = pd.to_datetime(df_la['time'])
df_la.set_index('time', inplace=True)
df_la.fillna(method='ffill', inplace=True)
```

```python
# Resample Los Angeles Data to Daily Averages

daily_avg_la = df_la.resample('D').mean()
```

```python
# Combine DataFrames for Comparison

combined_temp = pd.DataFrame({
    'New York': daily_avg['temperature_2m'],
    'Los Angeles': daily_avg_la['temperature_2m']
})
```

```python
combined_temp.head()
```

```python
# Plot Comparison of Daily Average Temperature

combined_temp.plot(kind='bar', figsize=(10, 6))
plt.title('Daily Average Temperature: New York vs Los Angeles')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Step 9: Advanced Analysis - Temperature Difference

# Calculate Temperature Difference
combined_temp['Temp Difference'] = combined_temp['Los Angeles'] - combined_temp['New York']
combined_temp
```

```python
# Step 10: Save Data to CSV (Optional)

# Save New York data
df.to_csv('new_york_weather.csv')

# Save Los Angeles data
df_la.to_csv('los_angeles_weather.csv')
```

```python
# Lets first get London Parameters and dataset ready for analysis

base_url = 'https://archive-api.open-meteo.com/v1/archive'

hourly_variables = ['precipitation']
latitude_london = 51.5074
longitude_london = -0.1278

params_london = {
    'latitude': latitude_london,
    'longitude': longitude_london,
   'start_date': '2022-01-01',
    'end_date': '2022-12-31',
    'hourly': ','.join(hourly_variables),
    'timezone': 'Europe/London'
}

response_london = requests.get(base_url, params=params_london)
```

```python
# error handeling checking response is 200 or not
if response_london.status_code == 200:
  print("Request successful!")
elif response_london.status_code == 400:
  print("Bad request. Check your parameters.")
else:
  print(f"Request failed with status code: {response_london.status_code}")
```

```python
# converting responded data in to dataframe for London

data = response_london.json()

df = pd.DataFrame(data['hourly'])
df.head()
```

```python
# checking for null value and data types

df.info()
```

```python
# convert time column in datatime formate and set it as index and rechecking for null values
df['time'] = pd.to_datetime(df['time'])
df.set_index('time', inplace=True)

df.isnull().sum()
```

```python
df.head()
```

```python
# clear any current plot from above codes
plt.clf()
```

```python
# plotting hourly precipitation in Londay for the whole year of 2022

plt.figure(figsize=(15,6))

plt.plot(df.index , df['precipitation'])

plt.title('Total Hourly Precipitation in London')

plt.xlabel('Date') 

plt.ylabel('Precipitation (mm)')

plt.show()
```

```python
# convert datafrem from hourly to total weekly format
weekly_avg = df[['precipitation']].resample('W').sum()
weekly_avg.head()
```

```python
#line plot of weekly Precipitation

plt.clf()
plt.figure(figsize=(14,6))
plt.plot(weekly_avg.index , weekly_avg['precipitation'])
plt.title('Total Weekly Precipitation in Seattle')
plt.xlabel('Week of Date')
plt.ylabel('Precipitation (mm)')
plt.show()
```

```python
# lets get Seattle Parameter and datafram ready

latitude_seattle = 47.6062
longitude_seattle = -122.3321

params_seattle = {
    'latitude': latitude_seattle,
    'longitude': longitude_seattle,
   'start_date': '2022-01-01',
    'end_date': '2022-12-31',
    'hourly': ','.join(hourly_variables),
    'timezone': 'America/Los_Angeles'
}

response_seattle = requests.get(base_url, params=params_seattle)
```

```python
# check response
if response_seattle.status_code == 200:
  print("Request successful!")
elif response_seattle.status_code == 400:
  print("Bad request. Check your parameters.")
else:
  print(f"Request failed with status code: {response_seattle.status_code}")
```

```python
# Prepare with simiplar steps as London
data_s = response_seattle.json()

dfs = pd.DataFrame(data_s['hourly'])

dfs['time'] = pd.to_datetime(dfs['time'])
dfs.set_index('time', inplace=True)
```

```python
# getting weekly precipitation data for seattle
weekly_avg_s = dfs[['precipitation']].resample('W').sum()
weekly_avg_s.head()
```

```python
# Combine DataFrames for Comparison

combined_precipitation = pd.DataFrame({
    'London': weekly_avg['precipitation'],
    'Seattle': weekly_avg_s['precipitation'],
})

```

```python
# let check the dataframe

combined_precipitation.head()
```

```python
#line plotting to compare

plt.clf()
plt.figure(figsize=(14,6))

combined_precipitation.plot(kind='line', figsize=(14, 6))

plt.title('Total Weekly Precipitation: London vs Seattle')

plt.xlabel('Week of Date')

plt.ylabel('Precipitation (mm)')

plt.xticks(rotation=45)

plt.grid(True)

plt.show()
```

```python
combined_precipitation.head()
```

```python
# lets get the cumulative total precipitation for both cities 

total_precipitation = combined_precipitation.cumsum()

total_precipitation.head()
```

```python
#line plotting to compare 

plt.clf()
plt.figure(figsize=(14,6))

total_precipitation.plot(kind='line', figsize=(14, 6))

plt.title('Total Precipitation to that week: London vs Seattle')

plt.xlabel('Month')

plt.ylabel('Cumulative Precipitation (mm)')

plt.xticks(rotation=45)

plt.grid(True)

plt.show()
```
